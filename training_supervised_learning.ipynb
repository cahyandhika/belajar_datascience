```python
import pandas as pd
```

```python
data = pd.read_csv("data/input/iris_data.csv")
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop('Species', axis = 1)
y = data['Species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5, metric="euclidean")
clf.fit(x_train, y_train)
```

```python
clf.predict(x_test)
```

```python
y_test
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.svm import SVC
```

```python
clf_svc = SVC(kernel="rbf", decision_function_shape="ovo")
```

```python
clf_svc.fit(x_train, y_train)
```

```python
clf_svc.predict(x_test)
```

```python
clf_svc.score(x_test, y_test)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf_nn = MLPClassifier(activation="relu", hidden_layer_sizes=(100,100,50),max_iter=1000)
```

```python
clf_nn.fit(x_train, y_train)
```

```python
clf_nn.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, clf.predict(x_test)))
```
